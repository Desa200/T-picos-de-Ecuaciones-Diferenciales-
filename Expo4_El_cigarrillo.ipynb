## El Problema del Cigarrillo

### *Matemáticas utilizadas:*
Ecuaciones diferenciales ordinarias lineales con coeficientes constantes.

*Comentario:* Un problema que involucra un sistema distribuido se formula en términos de una ecuación integral simple.




Estudiaremos el problema de la filtración a través de un cigarrillo encendido, lo cual es un problema de filtración con frontera móvil.

### Definiciones:

* Sistema distribuido:

 Este término se refiere a un sistema donde las propiedades, como concentración o temperatura, varían en diferentes puntos del espacio, no sólo en el tiempo. En este caso, la concentración del componente $Z$ (algún compuesto del humo) cambia a lo largo de la longitud del cigarrillo. Es decir, las características del sistema no son uniformes en todo el cigarrillo, sino que dependen de la posición.

* Ecuación integral simple:

Una ecuación integral es una forma de representar relaciones acumulativas en las que se calcula una cantidad (como la cantidad total de un componente absorbido a lo largo de la longitud del cigarrillo) como la suma (integración) de contribuciones pequeñas en diferentes posiciones. En este contexto, se usa una ecuación integral para describir cómo el componente $Z$ es absorbido a lo largo del cigarrillo.

* Filtración a través de un cigarrillo encendido:

El problema analiza cómo el humo y sus componentes pasan a través del cigarrillo mientras se inhala. Parte de estos componentes se absorben en el tabaco y en el filtro (si lo hay), y otra parte llega a los pulmones.

* Frontera móvil:

Aquí, la "frontera móvil" es el área de combustión en el cigarrillo que avanza a medida que se quema. Este borde cambia de posición conforme se fuma el cigarrillo, reduciendo la longitud disponible para la filtración, y afecta la dinámica de cómo los componentes son absorbidos en el tabaco.

Determinar cuánto y cómo pasan ciertas sustancias a través del cigarrillo permite identificar la exposición a compuestos específicos. Esto es fundamental para estudios de salud que buscan reducir los efectos perjudiciales del humo.

### 3. Modelado de efectos de salud
Entender la concentración y cantidad de sustancias inhaladas facilita modelar su impacto a largo plazo en el organismo. Es especialmente útil en la investigación de enfermedades relacionadas con el tabaco, como el cáncer y la enfermedad pulmonar obstructiva crónica (EPOC).

Como hipótesis tenemos que durante la inhalación estacionaria una fracción $a$ de cualquier sustancia $z$ del tabaco es constante. Esta sustancia se quema en la forma $\epsilon$ y es arrastrada a través del cigarro. Por otro lado, la fracción $1-a$ escapa al aire. Veamos más a detalle estas suposiciones:

## **Suposiciones para el modelo de filtración en el cigarrillo**

1. **Fracción constante $a$ del componente $Z$ en el tabaco que se quema**

   Durante una inhalación constante (es decir, a una velocidad constante de aire y de gases quemados que pasan a través del cigarrillo), y con una tasa constante de combustión, una fracción fija $a$ de cualquier componente $Z$ en el tabaco que se quema en la punta es aspirada a través del cigarrillo, mientras que la fracción restante, $1 - a$, se dispersa en el aire. En condiciones normales de fumar, el valor de $a$ suele estar cerca de la unidad, pero asumir un valor arbitrario de $a$ no complica esencialmente el análisis.

   Esta fracción $a$ representa una constante de transferencia, simplificando el análisis del sistema. La teoría detrás de esta suposición se basa en la *cinética de reacciones*, donde se modela cómo los materiales pasan por procesos de combustión y el transporte de compuestos en flujos de aire.

2. **Coeficientes de absorción del tabaco y del filtro**

   Respecto a la filtración del componente $Z$, el coeficiente de absorción del tabaco es constante y se denota por $b$. De manera similar, si hay una punta de filtro, el coeficiente de absorción de la punta es constante y se denota por $\beta$.

   Estos coeficientes, $b$ y $\beta$, representan la capacidad del tabaco y del filtro para retener el componente $Z$ mientras el humo pasa a través del cigarrillo. Matemáticamente, ambos coeficientes son constantes en el modelo y determinan la cantidad del componente que es atrapado en el material.

   Los valores $b$ y $\beta$ se basan en la *teoría de absorción*, aplicada en procesos de filtración para estudiar cómo ciertos materiales pueden retener o bloquear compuestos específicos. En este modelo, se considera que $b$ y $\beta$ permanecen constantes a lo largo del proceso.

3. **Suposición de tiempo de retardo despreciable**

   Se asume que la tasa de cambio de longitud del cigarrillo debido a la combustión es tan lenta que el retardo de tiempo asociado al flujo de gas a lo largo del cigarrillo es insignificante. En otras palabras, la deposición de la fracción $a$ de cualquier cantidad del componente $Z$ a lo largo del cigarrillo ocurre instantáneamente a medida que esta cantidad se quema en la punta.

   Esta suposición simplifica el modelo, permitiendo evitar cálculos de retardos y asumir que la combustión y la filtración suceden simultáneamente. Matemáticamente, esto significa que no se necesitan ecuaciones diferenciales que incluyan retardos, ya que el sistema se considera en estado *cuasi-estático*. En términos de la *teoría de sistemas dinámicos*, equivale a suponer un sistema con cambios de longitud muy lentos comparados con la velocidad de otros procesos involucrados.

Supongamos por simplicidad que el humo se mueve a lo largo del cigarrillo con velocidad constante, y consideremos una lámina delgada de grosor $\varepsilon$, digamos, moviéndose con el humo (Fig. 7.13). Sea el peso total del componente $Z$ en la lámina cuando pasa la posición $x$ sea $w(x)$. El peso total en $x + \Delta x$ es entonces $w(x + \Delta x)$. Además, por definición del coeficiente de absorción $b$, el peso removido de la lámina mientras se mueve de $x$ a $x + \Delta x$ es $bw \Delta x$. Por lo tanto,


$$
w(x) - w(x + \Delta x) = bw \Delta x.
$$

*Cambio en el peso del componente $Z$ entre $x$ y $x + \Delta x$:*

   La ecuación anterior describe la pérdida de peso de $Z$ en la lámina al pasar desde $x$ hasta $x + \Delta x$. La cantidad de $Z$ absorbida es proporcional a $bw \Delta x$, donde $b$ es el coeficiente de absorción. Esto representa un modelo de *absorción exponencial*: mientras más lejos viaja el humo, menos $Z$ queda en el humo porque más se ha "depositado".

   El coeficiente de absorción $b$ viene dado de la siguiente forma:

$$b=\frac{w(x)-w(x+\Delta x)}{w\Delta x}$$

Si dejamos que $\Delta x$ tienda a cero, obtenemos

$$bw=lim \Delta_{x \to 0} \frac{w(x)-w(x+\Delta x)}{w\Delta x}$$

así:


$$
\frac{dw}{dx} = -bw \quad \text{o} \quad w = w_0 e^{-bx},
$$

donde $w_0$ es el valor de $w$ en $x = 0$.

Esta es una *ecuación diferencial ordinaria (EDO) de primer orden* y se usa para modelar procesos de decaimiento exponencial, común en fenómenos de filtración y absorción. Físicamente, esta ecuación muestra cómo el peso del componente $Z$ disminuye exponencialmente con la distancia a medida que se absorbe a lo largo del cigarrillo. En teoría, es un caso particular de *modelos de absorción exponencial* que describe la reducción gradual de una sustancia conforme avanza en un medio.

Ya hemos notado que el peso depositado entre $x$ y $x + \Delta x$ es $bw \Delta x$; es decir, el peso absorbido por unidad de longitud en la posición $x$ está dado por

$$
bw_0 e^{-bx}.z
$$

Esta expresión representa la *densidad de absorción* a lo largo del cigarrillo y depende del coeficiente de absorción $b$ y de la cantidad inicial $w_0$ de $Z$. Es decir, la cantidad de $Z$ depositada disminuye a lo largo del cigarrillo de manera exponencial.

Si el cigarrillo se corta en $x = X$, es claro físicamente que esto no afecta el análisis desde $x = 0$ hasta $x = X$, ya que no hay reflejo desde el extremo $x = X$ y el humo pasa a través del plano $x = X$ de la misma manera, sin importar si hay un cigarrillo o filtro en $x > X$ o no. El peso del componente $Z$ que pasa a través del plano $x = X$ se puede encontrar calculando la cantidad que se habría depositado desde $x = X$ hasta el infinito si el cigarrillo hubiera sido de longitud infinita; es decir, de (7.28),

$$
\int_{X}^{\infty} bw_0 e^{-bx} \, dx = w_0 e^{-bX}.
$$

Este valor representa la cantidad de $Z$ que aún no ha sido absorbida y sigue avanzando. Matemáticamente, es el límite del proceso de absorción, considerando un cigarrillo de longitud infinita en esta teoría simplificada.

*Absorción en dos secciones con diferentes coeficientes:*

De manera similar, si el cigarrillo tiene un coeficiente de absorción $b$ en $0 \leq x \leq \xi$ y $\beta$ en $\xi \leq x \leq X$, en esta parte, se considera un cigarrillo con dos secciones, cada una con un coeficiente de absorción distinto, esta estructura refleja un cigarrillo con filtro, donde el filtro tiene una capacidad de absorción distinta al tabaco.
Así, en lugar de (7.29) tendríamos

$$ w =  \begin{cases}
w_0 e^{-bx}, & 0 \leq x \leq \xi, \\
W e^{-\beta x}, & \xi \leq x \leq X, \end{cases} $$

(si tiene filtro)

donde la constante $W$ se determina por la condición de que $w$ sea continua en $x = \xi$; es decir,

$$ w_0 e^{-b\xi} = W e^{-\beta \xi} $$ o

$$
w = w_0 e^{-b\xi  - \beta(x - \xi)}, \quad \xi \leq x \leq X.
$$


- **$w_0$**: Este es el peso inicial del componente $Z$ en la posición $x = 0$ (el inicio del cigarrillo).
- **$e^{-b\xi}$**: Representa el peso de $Z$ que queda después de que ha pasado una distancia $\xi$ en la sección del tabaco, donde $b$ es el coeficiente de absorción del tabaco. Cuanto mayor sea $b$, más peso se pierde debido a la absorción.
- **$W$**: Es el peso del componente $Z$ en la sección del filtro al llegar a $x = \xi$.
- **$e^{-\beta \xi}$**: Modela la cantidad de $Z$ que se queda en el filtro después de haber pasado una distancia $\xi$ en la sección del filtro, donde $\beta$ es el coeficiente de absorción del filtro.

- **$-\beta(x - \xi)$**: Modela la absorción de $Z$ dentro de la sección del filtro desde $\xi$ hasta $x$.


Estas ecuaciones reflejan que, aunque el cigarrillo tiene dos materiales con diferentes capacidades de absorción, el modelo sigue siendo válido al usar diferentes exponentes para cada sección. Este tipo de ecuación es común en *modelos por partes* en sistemas físicos donde hay cambios en el material o condiciones.

El peso depositado por unidad de longitud en la posición $x \ > \xi$ está dado por

$$
\beta w_0 e^{-b\xi - \beta(x - \xi)},
$$

y representa la cantidad que se absorbe en el filtro. En el modelo, esto depende de la capacidad de absorción del filtro ($\beta$) y de la posición dentro del filtro $(x - \xi)$.


y el peso que pasa por el plano $x = X$  (con filtro)

$$ w_0 e^{-b\xi - \beta(x - \xi)}. $$

Esta expresión muestra la cantidad de $Z$ que queda en el humo después de pasar por el filtro, lo cual es crucial para estimar cuánto llega efectivamente al fumador. Mientras que sin filtro tenemos:

$$w_0 e^{-bX}$$

La relación del peso del componente $Z$ que pasa por el plano $x = X$ cuando hay un filtro presente [(7.31)], con el que pasa cuando no hay filtro [(7.29)], es

$$z=\frac{w_0 e^{-bX}}{ w_0 e^{-b\xi - \beta(x - \xi)}}$$

$$ z=e^{-( \beta-b)(x - \xi)}. $$
donde $(x - \xi)$ es la longitud del filtro

Esta ecuación muestra cómo el filtro reduce la cantidad de $Z$ que pasa, dependiendo de la diferencia entre los coeficientes de absorción del tabaco y el filtro. Matemáticamente, esta diferencia exponencial representa la *efectividad del filtro*.



Ahora $W(x,t) = W(vt,t)$.

En el tiempo $\Delta t$ se quema una longitud $v \Delta t$, que contiene $v \Delta t W(vt,t)$ de la componente $z$, y se absorbió:

$$ a v \Delta t W(vt,t) $$

Y el depositado entre $x$ y $ \Delta x$ es:

$$ a b v t W(vt,t) e^{-b(x - vt)} \Delta x $$

El peso total depositado en el tiempo $t$ entre $x$ y $x + \Delta x$ es:
$$ a b v \int_{0}^{t} W(v \tau, \tau) e^{b(x - v \tau)} = W(x,t) \Delta x $$

Para obtener el peso total, agregamos $W(x,0) \Delta x$ y definimos:
$$
W(x,t) = W(x,0) + a b v \int_{0}^{t} W(v \tau, \tau) d \tau
$$

Si tenemos $x = v t$:
$$
\psi(t) = W(vt,0) e^{b vt} + a b v \int_{0}^{t} \psi(\tau) e^{b(t - \tau)} d \tau
$$

Esta es una ecuación integral.

Para el caso simple $W(x,0) = W_0$, c.t. inicial no depende de $x$.

Resolvemos $(x)$ derivando:
$$
\frac{d \psi}{d t} = b v W_0 e^{b v t} + a b v \psi
$$

Por factor integrante:
$$
\frac{d f}{d t} = \int b v W_0 e^{b v t} d t + a b v f d t
$$

Resolver con Transformada de Laplace, con la condición $W(x,0) = W_0$.

$$
f' - a b v f = b v W_0 e^{b v t}
$$

1. **Primera Ecuación: $W(x,t) = W(vt, t)$**
   - Esta ecuación indica que $W(x,t)$, una función que representa una cantidad en función de la posición $x$ y el tiempo $t$, depende del término $vt$, donde $v$ es una velocidad constante. Esto sugiere que la cantidad $W$ está siendo transportada a lo largo de la posición $x$ a una velocidad $v$.

2. **Cantidad quemada en un intervalo de tiempo $\Delta t$**
   - La longitud quemada durante un intervalo $\Delta t$ es $v \Delta t$. Si $W(vt, t)$ representa la densidad de la componente $z$ en esta longitud, entonces la cantidad quemada de la componente $z$ es:
     $$
     v \Delta t W(vt, t)
     $$

3. **Cantidad absorbida:**
   - La cantidad absorbida durante el mismo intervalo de tiempo es una fracción $a$ de la cantidad quemada, es decir:
     $$
     a v \Delta t W(vt, t)
     $$

4. **Cantidad depositada entre $x$ y $x + \Delta x$:**
   - La cantidad depositada entre $x$ y $\Delta x$ en el tiempo $t$ se expresa como:
     $$
     a b v t W(vt, t) e^{-b(x - vt)} \Delta x
     $$
   - Aquí, $a$ y $b$ son coeficientes de absorción y depósito, respectivamente. Este término representa una contribución al peso que decae exponencialmente con la distancia desde la posición $vt$ debido a la presencia del término exponencial $e^{-b(x - vt)}$.

5. **Peso total depositado en el tiempo $t$ entre $x$ y $x + \Delta x$:**
   - Para obtener el peso total depositado hasta el tiempo $t$ en el intervalo de posiciones $[x, x + \Delta x]$, integramos sobre el tiempo de $0$ a $t$:
     $$
     a b v \int_{0}^{t} W(v \tau, \tau) e^{b(x - v \tau)} d\tau = W(x,t) \Delta x
     $$
   - Esta integral acumula el depósito de peso a lo largo del tiempo, considerando la atenuación exponencial. La expresión resultante representa la cantidad de peso total en el intervalo $\Delta x$ en el instante $t$.

6. **Definición del peso total:**
   - El peso total $W(x,t)$ en función de su valor inicial $W(x,0)$ más el peso depositado hasta el tiempo $t$ es:
     $$
     W(x,t) = W(x,0) + a b v \int_{0}^{t} W(v \tau, \tau) d \tau
     $$
   - Aquí, $W(x,0)$ es la condición inicial de la función de peso en $x$, y el segundo término es la contribución adicional debido al depósito.

7. **Ecuación integral para el caso $x = vt$:**
   - Para el caso particular cuando $x = vt$, la función de peso en términos del tiempo es:
     $$
     \psi(t) = W(vt,0) e^{b vt} + a b v \int_{0}^{t} \psi(\tau) e^{b(t - \tau)} d \tau
     $$
   - Esto es una **ecuación integral** que describe cómo $\psi(t)$ evoluciona en el tiempo con el depósito acumulado.

8. **Resolución para una condición inicial constante $W(x,0) = W_0$:**
   - Si $W(x,0) = W_0$ (una constante), se deriva $\psi$ respecto al tiempo $t$:
     $$
     \frac{d \psi}{d t} = b v W_0 e^{b v t} + a b v \psi
     $$
   - Esta ecuación diferencial describe el cambio de $\psi(t)$ en el tiempo.

9. **Método del factor integrante:**
   - Aplicamos el método del factor integrante para resolver la ecuación diferencial:
     $$
     \frac{d f}{d t} = \int b v W_0 e^{b v t} \, d t + a b v f \, d t
     $$
   - Esto conduce a una solución para $f$ en función de $t$ al resolver la integral en términos de $t$.

10. **Uso de la Transformada de Laplace para simplificación:**
    - Dado que $W(x,0) = W_0$, podemos utilizar la **Transformada de Laplace** para resolver la ecuación diferencial:
      $$
      f' - a b v f = b v W_0 e^{b v t}
      $$
    - Esto facilita la resolución de la ecuación en el dominio de Laplace, donde los términos se vuelven más manejables.

## Conclusión sobre el Problema del Cigarrillo

El análisis del problema de filtración a través de un cigarrillo encendido permite comprender la dinámica de cómo los componentes nocivos del humo son absorbidos a medida que atraviesan el tabaco y, si está presente, el filtro. Usando ecuaciones diferenciales ordinarias y su representación como integrales, se modela la absorción exponencial de un componente $ Z $ del humo, mostrando que su cantidad disminuye a medida que se avanza en el cigarrillo.

La clave es que la absorción depende de coeficientes constantes ($ b $ y $ \beta $) que representan la capacidad del tabaco y del filtro, respectivamente. Este enfoque no solo ayuda a evaluar la eficacia de los filtros, sino que también es crucial para investigaciones sobre salud pública y toxicología, así como para optimizar el diseño de cigarrillos y filtros.

### Ejemplo Concreto

Supongamos que tenemos un cigarrillo con una longitud total de 10 cm. Si $ w_0 = 100 $ mg es la cantidad inicial del componente $ Z $, y suponemos $ b = 0.1 $ cm$^{-1}$ y $ \beta = 0.2 $ cm$^{-1}$ para la sección del filtro que comienza en 5 cm, podemos calcular cuánto del componente $ Z $ pasará por el plano $ x = 10 $ cm (el extremo del cigarrillo).

## Resolución

Para resolver el ejemplo de la absorción del componente $ Z $ a lo largo de un cigarrillo de 10 cm con las condiciones dadas, seguimos los pasos:

### Datos

- Longitud total del cigarrillo: $ X = 10 $ cm
- Cantidad inicial del componente $ Z $: $ w_0 = 100 $ mg
- Coeficiente de absorción del tabaco: $ b = 0.1 $ cm$^{-1}$
- Coeficiente de absorción del filtro: $ \beta = 0.2 $ cm$^{-1}$
- Inicio del filtro: $ \xi = 5 $ cm

### Funciones de Absorción

1. **Para $ 0 \leq x \leq \xi $** (donde solo hay tabaco):
   $$
   w(x) = w_0 e^{-bx}
   $$
2. **Para $ \xi < x \leq X $** (donde hay filtro):
   $$
   w(x) = w_0 e^{-b\xi} e^{-\beta(x - \xi)}
   $$

### Cálculos

1. **Absorción hasta $ x = 5 $ cm**:
   $$
   w(5) = 100 e^{-0.1 \cdot 5} = 100 e^{-0.5} \approx 100 \cdot 0.6065 \approx 60.65 \text{ mg}
   $$

2. **Absorción desde $ x = 5 $ cm hasta $ x = 10 $ cm**:
   Primero, calculamos $ W $ (la cantidad de $ Z $ que queda en $ x = 5 $):
   $$
   W = 100 e^{-0.5} \approx 60.65 \text{ mg}
   $$
   Luego, para $ x = 10 $:
   $$
   w(10) = W e^{-0.2 \cdot (10 - 5)} = 60.65 e^{-1} \approx 60.65 \cdot 0.3679 \approx 22.29 \text{ mg}
   $$


### Resumen de Resultados

- **Cantidad de $ Z $ en $ x = 5 $ cm**: aproximadamente $ 60.65 $ mg
- **Cantidad de $ Z $ en $ x = 10 $ cm**: aproximadamente $ 22.29 $ mg

Esto significa que, al final del cigarrillo, solo queda alrededor de $ 22.29 $ mg del componente $ Z $ en el humo que pasa por el extremo del cigarrillo.


### Simulación en Python

```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetros
w0 = 100  # mg, cantidad inicial
b = 0.1   # cm^-1, coeficiente de absorción del tabaco
beta = 0.2  # cm^-1, coeficiente de absorción del filtro
xi = 5  # cm, inicio del filtro
X = 10  # cm, longitud total del cigarrillo

# Función para calcular la cantidad de Z en función de la distancia
def cantidad_Z(x):
    if x <= xi:
        return w0 * np.exp(-b * x)
    else:
        W = w0 * np.exp(-b * xi)  # Continuidad en x = xi
        return W * np.exp(-beta * (x - xi))

# Distancias
distancias = np.linspace(0, X, 100)
cantidades = [cantidad_Z(x) for x in distancias]

# Gráfico
plt.plot(distancias, cantidades)
plt.title("Absorción del componente Z a lo largo del cigarrillo")
plt.xlabel("Distancia (cm)")
plt.ylabel("Cantidad de Z (mg)")
plt.grid()
plt.axvline(x=xi, color='red', linestyle='--', label='Inicio del Filtro')
plt.legend()
plt.show()
```
