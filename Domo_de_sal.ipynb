### Teorema de Taylor

El teorema de Taylor establece que cualquier funci贸n $ f(x) $ que sea suficientemente suave (es decir, que tenga derivadas continuas) puede ser aproximada por un polinomio en torno a un punto $ a $. La forma general del teorema es:

A continuaci贸n, necesitamos la ecuaci贸n de continuidad en coordenadas rectangulares, que es simplemente una declaraci贸n del hecho de que se supone que la sal es incompresible, o la cantidad neta de sal que cruza las seis superficies de un elemento rectangular con lados $x$, $y$ y $z$ es cero (Fig. 7.10). * Primera parte de la ecuaci贸n:

$$ (u_y + \frac{\delta u_y}{\delta y}) - u_y $$

Aqu铆, $ u_y $ es la velocidad de flujo en una cara del elemento rectangular y $ u_y + \frac{\delta u_y}{\delta y} $ es la velocidad en la otra cara. La diferencia entre estas dos cantidades es la **variaci贸n del flujo** entre ambas caras en la direcci贸n $ y $.

* **Producto por el 谩rea**:

Luego, esta variaci贸n de flujo se multiplica por el 谩rea de la superficie a trav茅s de la cual el flujo est谩 ocurriendo. El 谩rea de esta superficie es $ \Delta x \Delta z $, ya que est谩s mirando los lados paralelos al eje $ x $.

El flujo neto a trav茅s de los lados paralelos al eje $ x $ es entonces:

$$ \frac{\delta u_y}{\delta y} \Delta x \Delta y \Delta z $$

Esta es la tasa de cambio del flujo de sal en la direcci贸n $ y $, multiplicada por el volumen $ \Delta x \Delta y \Delta z $ del elemento rectangular.

* **Significado f铆sico**:

Esto indica que la cantidad neta de sal que entra y sale a trav茅s de un peque帽o volumen en el espacio debe ser cero si la sal es incompresible. La ecuaci贸n de continuidad se encarga de asegurar que la masa total dentro del volumen no cambia, lo que se expresa como la suma de los flujos en todas las direcciones siendo cero.

$$\frac{\delta u_x}{\delta x}+\frac{\delta u_y}{\delta y}+\frac{\delta u_z}{\delta z}=d_x+d_y+d_z=0$$


Usando este resultado en (7.21), e introduciendo la notaci贸n $a - b = 2\mu$, donde $\mu$ se conoce como el coeficiente de viscosidad, vemos que (7.21) se reduce a:

$$ _x= -p+2\mu \frac{\delta u_x}{\delta x},  _y= -p+2\mu \frac{\delta u_y}{\delta y}, _z= -p+2\mu \frac{\delta u_z}{\delta z} $$ (7.22)

Ahora interpretamos este resultado en coordenadas esf茅ricas. La tasa de deformaci贸n en la direcci贸n radial [comparar (7.19) y Fig. 7.11] es simplemente  $\frac{\delta u}{\delta r}$. La tasa de deformaci贸n en la direcci贸n tangencial est谩 dada por (Fig. 7.11)

$$\lim_{\theta \to 0} \frac{u sin\theta}{r\theta}= \frac{u}{r}$$

Utilizando estos resultados en (7.22), con $_x =_r$ , y $_y=_z=_t$, vemos

$$_r=-p+2\mu \frac{\delta u}{\delta r},  _t=-p+2\mu \frac{ u}{ r}$$

* $\frac{\delta u}{\delta r}$: es la derivada radial de la velocidad de deformaci贸n, que mide c贸mo cambia la velocidad con respecto al radio $r$.
