#### Ecuaciones Matemáticas

Con la aproximación adicional de la simetría esférica, se descartan las fuerzas inerciales, (las condiciones son las mismas en todas las direcciones desde el centro de la esfera), porque se considera un tipo de movimiento de arrastre.

 Denotemos la velocidad radial (positiva hacia afuera) por $u(r, t)$, donde $r$ es la distancia desde el centro de la cavidad hasta el punto en consideración y $t$ representa el tiempo.

1. **Condición de Continuidad**:

La condición de continuidad (incompresibilidad) establece que la cantidad de fluido que entra en un volumen debe ser igual a la cantidad que sale de ese volumen y que en cualquier instante la velocidad a la que la sal se mueve a través de la superficie de una esfera de radio $r$ es independiente de $r$; es decir,

   $$
   4 \pi r^2 u(r, t) = 4\pi (r+\Delta r)^2 u(r+\Delta r, t)
   $$

   **Desglose:**

- **$ 4\pi r^2 $**: Esta es el área de la superficie de una esfera de radio $ r $. Multiplicamos por la velocidad $ u(r,t) $ para obtener el flujo radial de sal a través de esa superficie.

- **$ u(r,t) $**: Representa la velocidad radial de la sal en un punto dado en el tiempo $ t $. La condición establece que este flujo debe ser constante en el tiempo y no depender del radio $ r $.

- **$ (r + \Delta r)^2 $**: Este término representa el área de la superficie de una esfera ligeramente más grande (de radio $ r + \Delta r $). La ecuación implica que el flujo a través de esta superficie también debe ser igual.

(Al analizar una esfera con un radio mayor $(r+Δr)$, se puede verificar que la cantidad de sal que entra y sale de un volumen debe ser la misma en cualquier instante. Esto es crucial para establecer la conservación de masa en un fluido incompresible.)

Si ambos flujos son iguales, se establece que la velocidad de la sal debe ajustarse en función del cambio en el área de la superficie, además nos permite introducir derivaciones.

   

2. **Ecuación Resultante**:

   O usando el teorema de Taylor y simplificando, se obtiene:

 $$[2ru(r, t) +r^2 \frac{\delta u(r, t)}{\delta r}] ∆r + O(∆r^2)=0$$

 #### Desglose:

1. **Teorema de Taylor**: Esto se usa para aproximar la función $u(r+\Delta r, t)$ en términos de $u(r, t)$ y su derivada.

---
### Teorema de Taylor

El teorema de Taylor establece que cualquier función $ f(x) $ que sea suficientemente suave (es decir, que tenga derivadas continuas) puede ser aproximada por un polinomio en torno a un punto $ a $. La forma general del teorema es:

$$
f(x) = f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f'''(a)}{3!}(x - a)^3 + \ldots + \frac{f^{(n)}(a)}{n!}(x - a)^n + R_n(x)
$$

donde:

- $ f'(a), f''(a), \ldots, f^{(n)}(a) $ son las derivadas de $ f $ evaluadas en el punto $ a $.
- $ n! $ es el factorial de $ n $.
- $ R_n(x) $ es el resto del teorema de Taylor, que mide el error de la aproximación y se puede expresar como:

$$
R_n(x) = \frac{f^{(n+1)}(c)}{(n+1)!}(x - a)^{n+1}
$$

para algún $ c $ entre $ a $ y $ x $.

### Uso

Este teorema es útil para aproximar funciones complicadas mediante polinomios, facilitando cálculos en análisis matemático.

---


2. **$2ru(r,t)$**: Este término representa la variación del flujo debido al cambio en el radio $r$.

3. **$\frac{\delta u(r,t)}{\delta r}$**: Es la tasa de cambio de la velocidad radial con respecto al radio. Si esto es diferente de cero, significa que **la velocidad no es constante** (por definición de derivada).

4. **$O(\Delta r^2)$**: Representa términos de orden superior, que se pueden despreciar cuando $\Delta r$ es pequeño.


 Al dividir por $\Delta r$ y tomar el límite cuando $\Delta r \to 0$, se obtiene que:


   $$
   \frac{\partial}{\partial r}(r^2 u) = 0 \implies u = \frac{f(t)}{r^2}
   $$

(esto último por la integración)

---

### Proceso de Obtención de la Ecuación

Partimos de la ecuación:

$$
[2ru(r,t) + r^2 \frac{\delta u(r,t)}{\delta r}] \Delta r + O(\Delta r^2) = 0
$$

#### Paso 1: Dividir por $\Delta r$

Al dividir toda la ecuación por $\Delta r$ (asumiendo que $\Delta r \neq 0$), obtenemos:

$$
2ru(r,t) + r^2 \frac{\delta u(r,t)}{\delta r} + O(\Delta r) = 0
$$

#### Paso 2: Tomar el Límite

A medida que $\Delta r$ tiende a cero, el término $O(\Delta r)$ también tiende a cero, por lo que podemos ignorarlo en el límite:

$$
2ru(r,t) + r^2 \frac{\delta u(r,t)}{\delta r} = 0
$$

#### Paso 3: Rearreglar la Ecuación

Reorganizamos la ecuación:

$$
r^2 \frac{\delta u(r,t)}{\delta r} = -2ru(r,t)
$$

#### Paso 4: Separar Variables

Dividimos ambos lados por $r^2$ (asumiendo $r \neq 0$):

$$
\frac{\delta u(r,t)}{\delta r} = -\frac{2u(r,t)}{r}
$$

#### Paso 5: Integrar

Esta es una ecuación diferencial de primer orden. Podemos integrar ambos lados respecto a $r$:

$$
\int \frac{1}{u(r,t)} \delta u = -2 \int \frac{1}{r} \delta r
$$

Lo que da:

$$
\ln(u) = -2 \ln(r) + C(t)
$$

donde $C(t)$ es una constante de integración que puede depender del tiempo.

#### Paso 6: Exponenciar

Al exponenciar ambos lados, obtenemos:

$$
u(r,t) = e^{C(t)} \cdot r^{-2} = \frac{f(t)}{r^2}
$$

donde $f(t) = e^{C(t)}$ es una función arbitraria del tiempo.

### Conclusión

Por lo tanto, al dividir por $\Delta r$ y tomar el límite cuando $\Delta r \to 0$, se llega a la conclusión de que:

$$
\frac{\partial}{\partial r}(r^2 u) = 0 \implies u = \frac{f(t)}{r^2}
$$

Esto indica que la velocidad radial de la sal depende del tiempo y es inversamente proporcional al cuadrado del radio.

Ahora consideramos algunas propiedades de las tensiones y distorsiones que deben establecerse (o aceptarse) antes de seguir avanzando.

La tensión en el medio se considera y se establece que no hay tensiones cortantes en la configuración esférica:

1. **Tensiones principales**: En un medio en equilibrio, las tensiones se dividen en tensiones radiales y tangenciales. Las tensiones radiales $\sigma_r$ actúan hacia fuera y son influenciadas por la presión y el movimiento del fluido.

2. **Viscosidad**: Se refiere a la resistencia al flujo. En el caso de un fluido newtoniano, las tensiones son proporcionales a la tasa de deformación.


La distorsión de un pequeño elemento de sal se ilustra en la figura 7.9, donde la línea continua representa una sección transversal del elemento en $t$ y la línea discontinua muestra la forma de este elemento en el tiempo $t+ \Delta t$.

Debido a la simetría esférica, no hay tensiones de corte en las superficies de coordenadas (es decir, paralelas a las superficies de cualquier esfera  $r=constante$, o en una dirección que se encuentra en un plano a través del centro de la esfera). Las únicas tensiones son las tensiones principales $\sigma_r$, y $\sigma_t$, donde $\sigma_r$, es la tensión de tracción en la dirección radial y $\sigma_t$, es la tensión tangencial en la dirección circunferencial. Estos componentes de tensión son funciones del radio y del tiempo.

Las fuerzas que se oponen al movimiento son de naturaleza viscosa. Para comprender mejor estas fuerzas, supongamos que un fluido se mueve a lo largo de una línea recta, con una velocidad $u(x)$ en la posición $x$. Si $u(x + ∆x)$, la velocidad en $x + ∆x$ es la misma que $u(x)$, el fluido simplemente se mueve como un todo y no se producen fuerzas internas o entre las capas del fluido.


Sin embargo, si

$$\lim_{∆x \to 0} \frac{u(x+∆x)- u(x)}{∆x} =\frac{du}{dx} =k \neq 0$$

(7.19)

(la velocidad del fluido no es constante y que hay un cambio en la velocidad conforme te mueves a lo largo de $x$. En términos físicos, esto indica que las diferentes capas del fluido se mueven a distintas velocidades, lo que provoca la formación de tensiones internas)

Entonces, la cantidad $k$ es una medida de la tasa de distorsión del fluido, conocida como tasa de deformación. Suponemos que la sal se comporta como un fluido newtoniano, en el que las fuerzas
viscosas son linealmente proporcionales a la tasa de deformación. Para derivar la forma precisa de la relación, nos apoyaremos en gran medida en un argumento de simetría. Para ello, trabajamos temporalmente en coordenadas rectangulares $x$, $y$ y $z$.

Solo necesitamos considerar el caso en el que no hay tensiones de corte presentes. Supongamos que $σ_x$, $σ_y$ y $σ_z$ son los tres componentes de la tensión normal, $u_x$, $u_y$ y $u_z$ son las velocidades correspondientes, y

$$d_x=\frac{\delta u_x}{\delta x},   d_y=\frac{\delta u_y}{\delta y},d_z=\frac{\delta u_z}{\delta z},$$

son las derivadas espaciales de las velocidades. (Estas derivadas representan cómo cambian las velocidades en función de las posiciones en las direcciones $x$, $y$ y $z$).
Cuando el material está en reposo, debe estar en un estado de presión hidrostática, (La presión hidrostática es la presión ejercida por un fluido en equilibrio, donde cada punto dentro del fluido experimenta la misma presión en todas las direcciones. Esto es crucial para asegurar que el material se mantenga en equilibrio sin deformaciones
). Denotando esta presión por $𝑝$, tenemos

$$σ_x= σ_y=σ_z=-p$$



Cuando la sal está en reposo, donde es necesario un signo negativo ya que las tensiones de tracción se consideran positivas, mientras que la presión negativa es compresiva. Estamos considerando una situación en la que no hay tensiones de corte (Fig. 7.10a). Cuando la sal está en movimiento, tenemos

$$σ_x= -p + f(d_x, d_y, d_z), \ \ \ σ_y= -p +g(d_x, d_y, d_z),  σ_z=-p+h(d_x, d_y, d_z)$$


 donde $f, g$ y $h$, las tensiones viscosas, se desvanecen cuando $d_x, d_y$ y $d_z$ se desvanecen.

**Tensiones Viscosas**: Las tensiones viscosas son una respuesta del material a la deformación, y su presencia indica que el material está fluyendo. A medida que las derivadas espaciales de las velocidades se desvanecen (es decir, cuando el flujo se detiene o es constante), las tensiones viscosas también se desvanecen, volviendo al estado de presión hidrostática.

- **Hipótesis Newtoniana**: La hipótesis newtoniana establece que la tensión viscosa es proporcional a la tasa de deformación. Es la suposición más simple y se ha confirmado en muchos casos prácticos, permitiendo predecir cómo un fluido responderá al movimiento y la presión.

Esta relación es fundamental para entender cómo un material se comporta tanto en reposo como en movimiento.

La hipótesis newtoniana (que es la hipótesis más simple posible, confirmada experimentalmente para muchos fluidos) es que la diferencia entre $σ_x$ y $-p$ es una función lineal de $d_x$, $d_y$ y $d_z$, y de manera similar para $σ_y$ y  $σ_z$

$$ σ_x=-p+a_{11}d_x+a_{12}d_y+a_{13}d_z,$$
$$ σ_y=-p+a_{21}d_x+a_{22}d_y+a_{23}d_z,$$
$$ σ_z=-p+a_{31}d_x+a_{32}d_y+a_{33}d_z,$$ (7.20)

 Ahora suponemos que el medio es isótropo (sus propiedades físicas son las mismas en todas las direcciones), lo que significa que las propiedades del medio no dependen de la dirección que estemos considerando. Si consideramos una distorsión en la dirección x, digamos, $d_x=1$ y $d_y= d_z= 0$, entonces las tensiones en las direcciones $y$ y $z$ son las mismas; es decir, de (7.20),

$$σ_y= -p+a_{21}=σ_z= -p+a_{31}$$ o $$a_{21}=a_{31}$$

similarmente,

$$a_{12}=a_{32},   a_{13}=a_{23} $$


La suposición de que las propiedades del medio no dependen de la dirección considerada en el medio también implica que dos estados de deformación, que difieren sólo por una rotación, producirán dos estados de tensión que difieren sólo por una rotación. Por lo tanto, si $d'_x=1$  y $d'_y=d'_z= 0$ producen tensiones $σ'_x=\alpha$ y $σ'_y=σ_z='\beta$, entonces $d''_y=1$ y $d''_x= d''_z= 0$, producirán tensiones$σ''_y=\alpha$ y $σ''_x=σ''_z='\beta$. Por lo tanto, de (7.20),


$$\alpha= -p+a_{11}=-p+a_{22}$$
$$\beta= a_{21}=a_{31} =a_{12}=a_{32}$$

Aplicando un argumento similar para una deformación en la dirección $z$ vemos que todos los coeficientes diagonales $a_{ii}$, son iguales, y todos los $a_{ij}$ fuera de la diagonal son iguales:
$$ a_{11}=a_{22}=a_{33}=a,$$
$$a_{12}=a_{13}= a_{21}=a_{23}=a_{31}=a_{32}=b,$$

 donde estas ecuaciones definen $a$ y $b$.

 Introduciendo estos resultados, las ecuaciones (7.20) se convierten en
$$ σ_x=-p+ad_x+b(d_y+ d_z)$$
$$ σ_y=-p+ad_y+b(d_x+ d_z)$$
$$ σ_z=-p+ad_z+b(d_x+ d_y)$$

(7.21)


A continuación, necesitamos la ecuación de continuidad en coordenadas rectangulares, que es simplemente una declaración del hecho de que se supone que la sal es incompresible, o la cantidad neta de sal que cruza las seis superficies de un elemento rectangular con lados $∆x$, $∆y$ y $∆z$ es cero (Fig. 7.10). La tasa neta de flujo a través de los lados paralelos al eje $x$ es:

...........................



$$[(u_y+\frac{\delta u_y}{\delta y})-u_y]∆x∆z=\frac{\delta u_y}{\delta y}∆x∆y∆z$$

* Primera parte de la ecuación:

$$ (u_y + \frac{\delta u_y}{\delta y}) - u_y $$

Aquí, $ u_y $ es la velocidad de flujo en una cara del elemento rectangular y $ u_y + \frac{\delta u_y}{\delta y} $ es la velocidad en la otra cara. La diferencia entre estas dos cantidades es la **variación del flujo** entre ambas caras en la dirección $ y $.

* **Producto por el área**:

Luego, esta variación de flujo se multiplica por el área de la superficie a través de la cual el flujo está ocurriendo. El área de esta superficie es $ \Delta x \Delta z $, ya que estás mirando los lados paralelos al eje $ x $.

* **Resultado final**:

El flujo neto a través de los lados paralelos al eje $ x $ es entonces:

$$ \frac{\delta u_y}{\delta y} \Delta x \Delta y \Delta z $$

Esta es la tasa de cambio del flujo de sal en la dirección $ y $, multiplicada por el volumen $ \Delta x \Delta y \Delta z $ del elemento rectangular.

* **Significado físico**:

Esto indica que la cantidad neta de sal que entra y sale a través de un pequeño volumen en el espacio debe ser cero si la sal es incompresible. La ecuación de continuidad se encarga de asegurar que la masa total dentro del volumen no cambia, lo que se expresa como la suma de los flujos en todas las direcciones siendo cero.

---

La suma de esta contribución y contribuciones similares de los lados paralelos al eje $y$ y al eje $z$ da:


$$\frac{\delta u_x}{\delta x}+\frac{\delta u_y}{\delta y}+\frac{\delta u_z}{\delta z}=d_x+d_y+d_z=0$$


Usando este resultado en (7.21), e introduciendo la notación $a - b = 2\mu$, donde $\mu$ se conoce como el coeficiente de viscosidad, vemos que (7.21) se reduce a:

$$ σ_x= -p+2\mu \frac{\delta u_x}{\delta x},  σ_y= -p+2\mu \frac{\delta u_y}{\delta y}, σ_z= -p+2\mu \frac{\delta u_z}{\delta z} $$ (7.22)

Ahora interpretamos este resultado en coordenadas esféricas. La tasa de deformación en la dirección radial [comparar (7.19) y Fig. 7.11] es simplemente  $\frac{\delta u}{\delta r}$. La tasa de deformación en la dirección tangencial está dada por (Fig. 7.11)

$$\lim_{∆\theta \to 0} \frac{u sin∆\theta}{r∆\theta}= \frac{u}{r}$$

Utilizando estos resultados en (7.22), con $σ_x =σ_r$ , y $σ_y=σ_z=σ_t$, vemos

$$σ_r=-p+2\mu \frac{\delta u}{\delta r},  σ_t=-p+2\mu \frac{ u}{ r}$$


Aquí:

* $\sigma_r$: es la tensión radial, o sea, la fuerza por unidad de área que actúa en la dirección radial (hacia el centro de la esfera).

* $\sigma_t$: es la tensión tangencial, que es la fuerza por unidad de área que actúa en las direcciones perpendiculares al radio, o tangentes a la esfera.
$p$: representa la presión hidrostática que actúa en la sal.

* $\mu$: es el módulo de viscosidad de la sal, un parámetro que describe la resistencia al flujo del material.

* $u$: es la componente radial de la velocidad de deformación de la sal.

* $\frac{\delta u}{\delta r}$: es la derivada radial de la velocidad de deformación, que mide cómo cambia la velocidad con respecto al radio $r$.



 Como la velocidad de movimiento es extremadamente lenta, podemos suponer que, en cualquier instante, los efectos dinámicos son despreciables (como aceleraciones) y necesitamos considerar sólo las condiciones de equilibrio estático.

 Consideramos el equilibrio de un pequeño trozo de sal entre el radio $r$ y $r + ∆r$, que se muestra gráficamente en la figura 7.12a, donde el plano OABCD se vuelve a dibujar en detalle en la figura 7.12b. El elemento de sal tiene seis caras, dadas por partes de dos esferas de radios $r$ y $r + Δr$, y cuatro planos que se cortan entre sí en ángulos rectos, siendo los ángulos entre planos opuestos $\Delta \alpha$ (véase la figura 7.12 b). Las áreas de las superficies curvas de radios $r$ y $r + \Delta r$ son $(r Δα)^2$ y $[(r + \Delta r) Δα]^2$, con fuerzas por unidad de área de $\sigma_r(r)$ y $\sigma_r (r + \Delta r)$, respectivamente. (Recuerde que se supone que la situación física es esféricamente simétrica, por lo que $\sigma_r$, y $\sigma_t$, dependen sólo de $r$ y $t$)

La fuerza neta en una esfera hacia afuera debida a las fuerzas sobre la superficie curva de radios $ r $ y $ r + \Delta r $ está dada por

$$
[(r + \Delta r) \Delta \alpha]^2 \sigma_r (r + \Delta r) - (r \Delta \alpha)^2 \sigma_r(r) = (r^2 + 2r \Delta r) (\Delta \alpha)^2 \left[ \sigma_r(r) + \frac{\delta \rho_r}{\delta r} \Delta r \right] - r^2 (\Delta \alpha)^2 \rho_r(r) $$+ término de orden superior



$$
= (2r \rho_r + r^2 \frac{\delta \rho_r}{\delta r}) \Delta r (\Delta \alpha)^2 + \text{términos de orden superior.} \tag{7.24}
$$

La fuerza en un punto del lado ABCD es $ \sigma _t(r + \xi \Delta r) $ por unidad de área, donde $ 0 \leq \xi \leq 1 $, el valor de $ \xi $ depende del punto preciso donde se mide la fuerza. Esta fuerza forma un ángulo $ (\frac{1}{2}\pi -\frac{1}{2} \Delta \alpha )$ con la línea OP en la figura 7.12a, donde O es el centro de la esfera y P es el centro de la cara curva de radio $ r $. El componente en la dirección OP de la fuerza en cualquier punto de la cara ABCD es

$$
-\sigma_t(r + \xi \Delta r) \sin \left(\frac{1}{2} \Delta \alpha \right) = -\frac{1}{2} \Delta \alpha \sigma_t(r) + \text{términos de orden superior.} \tag{7.25}
$$

La contribución total a la fuerza en la dirección OP se obtiene multiplicando esto por el área de ABCD, que es $ r \Delta r \Delta \alpha$, y por un factor de 4 ya que hay cuatro lados planos. De (7.24) y (7.25), el equilibrio de fuerzas en la dirección radial da, despreciando los términos de orden superior,

$$
(2r \sigma_r + r^2 \frac{\delta \sigma_r}{\delta r}) \Delta r (\Delta \alpha)^2 - 4r \Delta r \Delta \alpha \left(\frac{1}{2} \Delta \alpha\right) \sigma_t = 0 \tag{7.26}
$$

o

$$
\frac{\delta \sigma_r}{\delta r} + \frac{2(\sigma_r - \sigma_t)}{r} = 0.
$$

Las ecuaciones básicas son (7.18), (7.23) y (7.26). Sustituyendo (7.18) y (7.23) en (7.26), obtenemos $ \frac{\delta p}{\delta r} = 0 $, por lo tanto $ p = p(t) $, independiente de $ r $. Estamos asumiendo que la cavidad está en una región infinita en la que la tensión en el infinito no se ve afectada por la cavidad. Por lo tanto, $ p $ es constante, independiente de $ r $ y $ t $. Sea el radio de la cavidad en los momentos $ a(t) $. En la superficie de la cavidad, $ p = 0 $. Por lo tanto, la primera ecuación en (7.23), junto con (7.18), da

$$
f(t) = -\frac{p a^3}{4\mu}.
$$



**Relación de Velocidad**:
   Al integrar, se obtiene que la velocidad de cierre de la cavidad depende del tiempo y el radio:
   $$
   u(a, t) = -\frac{pa}{4\mu}
   $$
   donde $ p $ es la presión y $ \mu $ la viscosidad.

   Pero la velocidad de la superficie de la cavidad es da/dt, por lo que da

$$\frac{da}{dt}=u(a,t)=-\frac{pa}{4\mu}$$


#### Ecuación de Cierre de Cavidades

Integrando la relación de velocidad (pues p es una constante):

$$
\frac{da}{dt} = -\frac{pa}{4\mu} \implies a(t) = a_0 e^{-\frac{pt}{4\mu}}
$$
donde $ a_0 $ es el radio inicial.

**Interpretación**:

 Esto implica que la tasa de cambio del radio de la cavidad es proporcional a la presión y se ve afectada por la viscosidad. Un aumento en la viscosidad lleva a un cierre más lento de la cavidad.

El tiempo $ T $ en el que el radio de la cavidad se reduce en un factor $ 1/e=0,368 $ está determinado por $pT/4μ = 1$; es decir:
$$
T = \frac{4\mu}{p}
$$
Este es el tiempo de relajación, que aumenta con la viscosidad y disminuye con la presión.

 **Tiempo de relajación**: El tiempo que toma para que el radio de la cavidad se reduzca en un factor específico es inversamente proporcional a la presión y directamente proporcional a la viscosidad, lo que tiene implicaciones prácticas en la modelación de la sal y su comportamiento en condiciones específicas.

Como era de esperar, aumenta con la viscosidad $\mu$ y disminuye con la presión circundante p. La presión p es más o menos proporcional a la profundidad. La viscosidad de la sal en las condiciones existentes en un domo de sal debe determinarse experimentalmente. Se podrían realizar experimentos para comprobar la validez de la teoría anterior en el campo midiendo el cambio con el tiempo del volumen de una cavidad en un domo de sal.


### Conclusión

Estos conceptos y fórmulas describen un sistema físico complejo donde la viscosidad y la presión juegan roles críticos en el comportamiento de un fluido incompresible, en este caso, la sal. Se usan principios de mecánica de fluidos y teoría de tensiones para entender cómo se comporta el sistema en diferentes condiciones.

### Ejemplo

Imaginemos una cavidad con un radio inicial de 10 m en un domo de sal, a una presión de 2000 Pa y con una viscosidad de 0.5 Pa·s.

- **Calcular el tiempo de relajación**:
  $$
  T = \frac{4 \cdot 0.5}{2000} = 0.001 \text{ s}
  $$
  
Esto sugiere que la cavidad se cerraría rápidamente bajo estas condiciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
a0 = 10  # Radio inicial de la cavidad (m)
p = 2000  # Presión circundante (Pa)
mu = 0.5  # Viscosidad de la sal (Pa·s)

# Tiempo de relajación
T_relajacion = 4 * mu / p
print(f"Tiempo de relajación: {T_relajacion:.6f} segundos")

# Función para calcular el radio de la cavidad a lo largo del tiempo
def radio_cavidad(t, a0, p, mu):
    return a0 * np.exp(-p * t / (4 * mu))

# Tiempo en el que se evaluará la simulación (hasta 10 veces el tiempo de relajación)
tiempo = np.linspace(0, 10 * T_relajacion, 1000)

# Cálculo del radio en cada instante de tiempo
radio = radio_cavidad(tiempo, a0, p, mu)

# Graficar el resultado
plt.figure(figsize=(8, 6))
plt.plot(tiempo, radio, label='Radio de la cavidad')
plt.axvline(x=T_relajacion, color='r', linestyle='--', label=f'Tiempo de relajación ({T_relajacion:.6f} s)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Radio de la cavidad (m)')
plt.title('Cierre de la cavidad en un domo de sal')
plt.grid(True)
plt.legend()
plt.show()
```
