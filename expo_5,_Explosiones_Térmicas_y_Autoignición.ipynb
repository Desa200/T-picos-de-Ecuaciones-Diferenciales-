Problema de Transferencia de Calor:
Consideramos un cilindro infinito que genera calor a una tasa uniforme, q. La temperatura u(r) en función de la distancia radial r es lo que queremos determinar.
Este problema se modela matemáticamente mediante una ecuación diferencial que describe cómo se distribuye la temperatura en el cilindro.


Consideremos un cuerpo en el que una reacción libera calor en todo el volumen a una tasa que aumenta con la temperatura. Además, el calor se pierde de la superficie del cuerpo a una tasa proporcional a la diferencia entre la temperatura de la superficie y la temperatura circundante. Son posibles dos tipos de comportamiento:



1. En el primer caso, se establece un equilibrio con un diferencial de temperatura relativamente pequeño, lo que permite conducir lejos el calor liberado.

(la temperatura se mantiene constante).

Por ejemplo, un cilindro de metal que se calienta de manera uniforme debido a una resistencia eléctrica en su interior. La temperatura de la resistencia genera calor, que se distribuye en el cilindro.

Un sistema puede alcanzar un equilibrio térmico con un diferencial de temperatura relativamente pequeño gracias a la eficiente transferencia de calor en materiales conductores, donde el calor generado se distribuye rápidamente. Además, la convección y radiación en la superficie del sistema ayudan a perder calor de manera efectiva, lo que contribuye a mantener diferencias de temperatura mínimas.





2. En el segundo caso, el calor no se conduce lo suficientemente rápido y la temperatura interior se eleva a un valor alto.

 ( la temperatura aumenta significativamente)

Imaginemos un gran tanque de agua en el que se coloca un calentador eléctrico. Si el calentador genera calor a una tasa muy alta, pero el agua tiene una baja conductividad térmica, el calor no se distribuye rápidamente por todo el volumen del agua. Esto provoca que la temperatura en la zona cercana al calentador se eleve significativamente, mientras que las áreas más alejadas permanecen frías. Como resultado, se establece un fuerte diferencial de temperatura, y la temperatura interior del tanque puede alcanzar niveles peligrosamente altos, aumentando el riesgo de ebullición o incluso explosión. Esto sucede porque la tasa de generación de calor supera la capacidad del sistema para disiparlo, creando un desequilibrio térmico.




En la mayoría de los casos, la transición entre los dos tipos de soluciones es abrupta, (incluso un pequeño cambio en las condiciones, como un aumento en la tasa de generación de calor o un cambio en las propiedades del material, puede llevar a un colapso del equilibrio térmico), y el objetivo de la matemática es mostrar cómo es posible esto. Obviamente, un aumento en el tamaño favorece el segundo tipo de solución, debido a la menor relación entre superficie y volumen.

> Describimos ahora una situación idealizada en la que el fenómeno puede ser examinado matemáticamente. Imaginamos un cilindro infinito en el que se está liberando calor y queremos encontrar bajo qué condiciones su temperatura tiende a un valor finito conforme el tiempo tiende a infinito, en contraposición al caso en que la temperatura se aproxima al infinito. (En el segundo caso, el modelo matemático colapsará en un tiempo finito debido a una explosión o combustión).


 La pregunta es ¿cómo se comporta la temperatura en este cilindro con el tiempo?. Se busca saber si, después de mucho tiempo, la temperatura llegará a un valor finito estable o si continuará aumentando sin límite, lo cual llevaría a un colapso del modelo por fenómenos como explosión o combustión incontrolada.


> Como se había hablado en exposiciones anteriores, considerar un cilindro infinito permite despreciar los efectos de borde que ocurren en sistemas finitos. Además, al examinar cómo la temperatura se comporta a medida que el tiempo tiende a infinito, se pueden identificar condiciones de equilibrio y determinar si la temperatura se estabiliza en un valor finito o si aumenta indefinidamente.

 Esto permite concentrarse únicamente en cómo varía la temperatura en función del radio y del tiempo, sin preocuparse por el flujo de calor en los extremos.


> Abordaremos el problema de dos formas distintas: Se utilizarán dos enfoques: (1) una solución exacta mediante una ecuación diferencial parcial (EDP) no lineal, denominada como ecuación (9.19), y (2) un método aproximado basado en técnicas previas. La comparación entre ambos enfoques permitirá verificar la consistencia y exactitud del modelo.

> Para simplificar, si el cilindro infinito está en el rango $ 0 \leq r \leq a $, $ -\infty < z < \infty $ (donde $ z $ es la coordenada a lo largo del eje longitudinal del cilindro, es decir que no varían los parámetros a lo largo del cilindro), asumimos que todas las cantidades son independientes de $ z $, de modo que la temperatura depende solo de $ r $ y $ t $. Supongamos que el calor se libera de acuerdo con la ley de Arrhenius, es decir, a una tasa $ Q e^{-P/T} $ por unidad de volumen, donde $ Q $ es una constante que representa la energía liberada por unidad de volumen y $ P $ es una constante que depende de la reacción química que libera calor y $ T $ es la temperatura absoluta.

 Se asume que el cilindro es homogéneo a lo largo del eje $ z $ (es decir, todas las propiedades son independientes de $ z $). Esto reduce la dependencia espacial de la temperatura a las coordenadas radiales $ r $ y el tiempo $ t $.

 La expresión depende de $ T $, indicando que la tasa de liberación de calor disminuye exponencialmente a medida que aumenta $ T $.

> Si la temperatura en el cilindro a un radio $ r $ y en un tiempo $ t $ es denotada por $ T(r, t) $, la ecuación para $ T $ se puede formular considerando el balance de calor en un anillo que se encuentra entre $ r $ y $ r + \Delta r $, de la siguiente manera (ver Figura 9.7).


Aquí se plantea la configuración matemática para obtener la ecuación diferencial de la temperatura. Se considera un "anillo" de material en el cilindro, con un espesor $ \Delta r $, en el cual se analizará el balance de calor. Este enfoque permite derivar una ecuación diferencial que describe cómo cambia la temperatura $ T(r, t) $ en el tiempo y en el radio, al considerar el flujo de calor a través de las fronteras del anillo y el calor producido dentro de él.

> Consideremos el intervalo de tiempo de $ t $ a $ t + \Delta t $. Entonces,

$$
 \text{[Incremento de calor en el contenido del anillo]} = \text{[Calor neto entrando en el anillo a través de sus superficies]} + \text{[Calor producido por la reacción]}.
 $$

Este balance térmico establece la relación entre tres términos en un intervalo de tiempo pequeño $ \Delta t $:
1. *Incremento de calor en el anillo:* Representa el cambio de energía térmica dentro del anillo.
2. *Calor neto entrando en el anillo:* Es la cantidad de calor que fluye a través de las superficies interna y externa del anillo.
3. *Calor producido por la reacción:* Es la cantidad de calor generada en el volumen del anillo debido a la reacción exotérmica.

> En símbolos, omitiendo $ \Delta t $ en todo momento,
>
> $$
> \rho c \frac{\partial T}{\partial t} 2 \pi r \Delta r = k \left[ 2 \pi (r + \Delta r) \frac{\partial T(r + \Delta r, t)}{\partial r} - 2 \pi r \frac{\partial T(r, t)}{\partial r} \right] + 2 \pi r \Delta r Q e^{-P/T}.
> $$
Aquí se escribe el balance térmico en forma matemática. Cada término en la ecuación representa:

- $ \rho c \frac{\partial T}{\partial t} 2 \pi r \Delta r $: Es el cambio de energía en el volumen del anillo. $ \rho $ es la densidad del material, $ c $ es su capacidad calorífica específica, y $ \frac{\partial T}{\partial t} $ es la tasa de cambio de la temperatura.
- $ k \left[ 2 \pi (r + \Delta r) \frac{\partial T(r + \Delta r, t)}{\partial r} - 2 \pi r \frac{\partial T(r, t)}{\partial r} \right] $: Representa el flujo neto de calor a través de las superficies del anillo. $ k $ es la conductividad térmica del material.
- $ 2 \pi r \Delta r Q e^{-P/T} $: Es el calor generado en el anillo por la reacción química exotérmica.
> Dividiendo por $ \Delta r $ y dejando que $ \Delta r $ tienda a cero, encontramos que
>
> $$
> \rho c \frac{\partial T}{\partial t} = k \left( \frac{\partial^2 T}{\partial r^2} + \frac{1}{r} \frac{\partial T}{\partial r} \right) + Q e^{-P/T}.
> $$(9.17)

Al dividir ambos lados de la ecuación anterior por $ \Delta r $ y hacer $ \Delta r \to 0 $, se obtiene una ecuación diferencial parcial (EDP) que describe la variación de la temperatura $ T(r, t) $ en función del tiempo y del radio:

- *Lado izquierdo:* $ \rho c \frac{\partial T}{\partial t} $ representa el cambio de energía térmica en el anillo.
- *Primer término en el lado derecho:* $ k \left( \frac{\partial^2 T}{\partial r^2} + \frac{1}{r} \frac{\partial T}{\partial r} \right) $ describe la difusión de calor en el cilindro.
- *Segundo término en el lado derecho:* $ Q e^{-P/T} $ representa el calor generado por la reacción exotérmica

Esta es una ecuación diferencial parcial no lineal que debe resolverse para $T$ en $0 \leq r \leq a$, con condiciones iniciales adecuadas en $t = 0$ y condiciones de contorno en $r = a$. Supondremos, de hecho, que el límite $r = a$ se mantiene a una temperatura constante
$$
T = T_0 \quad \text{en} \quad r = a, \, t > 0.
$$
(9.18)

Haremos la suposición simplificadora de que $T$ no difiere mucho de $T_0$, por lo que podemos escribir, aproximadamente,

$$
\frac{1}{T} = \frac{1}{T_0} \left( 1 + \frac{T - T_0}{T_0} \right)^{-1} \approx \frac{1}{T_0} - \frac{T - T_0}{T_0^2}.
$$

Esta es una __aproximación lineal__ que se hace para tratar de simplificar una expresión cuando $T$ (la temperatura variable) está cerca de $T_0$ (una temperatura de referencia). Se usa una expansión binomial de primer orden:

La expresión inicial es:

$$
\frac{1}{T} = \frac{1}{T_0} \left( 1 + \frac{T - T_0}{T_0} \right)^{-1}.
$$

Aquí se está utilizando la relación entre $T$ y $T_0$ en términos de un pequeño cambio $\frac{T - T_0}{T_0}$. Como $T$ es cercano a $T_0$, podemos usar la expansión binomial para $\left( 1 + x \right)^{-1}$, donde $x = \frac{T - T_0}{T_0}$, que da como resultado:

$$
\left( 1 + \frac{T - T_0}{T_0} \right)^{-1} \approx 1 - \frac{T - T_0}{T_0}.
$$

Entonces,
$$
\frac{1}{T} \approx \frac{1}{T_0} - \frac{T - T_0}{T_0^2}.
$$
Así,

$$
Q e^{-P/T} \approx Q e^{-P/T_0} e^{P(T - T_0)/T_0^2}.
$$


Aquí se está aplicando la __aproximación de temperatura__ obtenida previamente en una expresión más compleja.

Se utiliza la forma exponencial de $e^{-P/T}$. Usando la aproximación anterior de $\frac{1}{T}$, tenemos:

$$
e^{-P/T} \approx e^{-P/T_0} e^{P(T - T_0)/T_0^2}.
$$

Esta es una expansión en términos de $T - T_0$, que es pequeña. La propiedad de los exponentes $e^{a + b} = e^a e^b$ se usa para separar las dos exponenciales.

__Explicación física:__

Esta es una forma común de aproximar funciones exponenciales en situaciones donde $T$ varía solo ligeramente con respecto a $T_0$. La ecuación representa una modificación de un factor $Q$ que depende exponencialmente de la temperatura, como se podría esperar en procesos de difusión o reacciones químicas dependientes de la temperatura.



Si establecemos

$$
T - T_0 = u, \quad \frac{1}{k} Q e^{-P/T_0} = q, \quad \frac{P}{T_0^2} = p,
$$

las ecuaciones (9.17) y (9.18) se convierten en

$$
\frac{1}{\kappa} \frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial r^2} + \frac{1}{r} \frac{\partial u}{\partial r} + q e^{pu}, \quad \kappa = \frac{k}{\rho c},
$$

(9.19)

$$
u = 0 \quad \text{en} \quad r = a.
$$


Se definen nuevas variables y parámetros para simplificar las ecuaciones posteriores:

- $u = T - T_0$ es el __desplazamiento de temperatura__ respecto a la temperatura de referencia $T_0$.
- $q = \frac{1}{k} Q e^{-P/T_0}$ es una constante que agrupa varios factores.
- $p = \frac{P}{T_0^2}$ es otro parámetro que involucra $P$ y $T_0$.

La ecuación (9.19) es una __ecuación de difusión con fuentes no lineales__. Se trata de una ecuación diferencial parcial (EDP) que describe cómo evoluciona $u$ (la diferencia de temperatura) en función del tiempo $t$ y la posición radial $r$ en un sistema (probablemente un sistema térmico esférico). El término $q e^{pu}$ representa una fuente no lineal dependiente de $u$.

- $\kappa = \frac{k}{\rho c}$ es la __difusividad térmica__, que involucra la conductividad térmica $k$, la densidad $\rho$, y la capacidad calorífica $c$.

__Explicación física:__

La ecuación (9.19) describe un proceso térmico en un medio (probablemente un sólido), donde la temperatura cambia con el tiempo y la posición radial. La presencia del término $e^{pu}$ sugiere que el proceso tiene una fuente térmica dependiente de la temperatura local.

La condición $u = 0$ en $r = a$ es una __condición de frontera__ que implica que la diferencia de temperatura $u$ se anula en el radio $r = a$.




Si $u$ tiende a un valor límite cuando $t$ tiende a infinito, es posible encontrar este valor límite de la siguiente manera. La cantidad $\partial u / \partial t$ en (9.19) tenderá a cero para $t$ grande, y el valor límite de $u$, si existe, satisfará

$$
\frac{d^2 u}{dr^2} + \frac{1}{r} \frac{du}{dr} + q e^{pu} = 0.
$$

(9.20)

Cuando $t \to \infty$, la derivada temporal $\frac{\partial u}{\partial t}$ tiende a cero, lo que significa que el sistema ha alcanzado el __estado estacionario__. Esto convierte la ecuación en una ecuación diferencial de segundo orden en $r$, donde la evolución de la temperatura $u$ depende únicamente de la variable radial.

__Explicación física:__

El sistema térmico llega a un estado en el que la temperatura no cambia con el tiempo, sino solo con la posición radial $r$. La ecuación resultante describe el __perfil de temperatura estacionario__ en el sistema.

Esta ecuación se puede resolver de la siguiente manera. Introduzcamos

$$
v = r \frac{du}{dr}
$$

(9.21)

y usemos esto para escribir (9.20) en la forma

$$
\frac{1}{r} \frac{dv}{dr} = -q e^{pu}.
$$

(9.22)

Se introduce la variable auxiliar $v = r \frac{du}{dr}$, lo que permite simplificar la ecuación diferencial y expresarla en una forma más manejable. Esto es un __cambio de variable__ que transforma la ecuación original en una ecuación de primer orden respecto a $v$.

La ecuación (9.22) es más fácil de resolver que la original.

La diferenciación de esta ecuación da

$$
\frac{d}{dr} \left( \frac{1}{r} \frac{dv}{dr} \right) = -qpe^{pu} \frac{du}{dr} = -qpe^{pu} \frac{v}{r}.
$$

(9.23)

Se toma la derivada de la ecuación (9.22) respecto a $r$, lo que da como resultado una nueva expresión para la evolución de $v$.

__Explicación física:__

La ecuación obtenida describe cómo cambia $v$ con respecto a la variable radial $r$.

Eliminación de $e^{pu}$ entre (9.22) y (9.23) produce

$$
\frac{d}{dr} \left( \frac{1}{r} \frac{dv}{dr} \right) = \frac{pv}{r^2} \frac{dv}{dr}.
$$

Este paso es una manipulación algebraica de las ecuaciones (9.22) y (9.23). El término $e^{pu}$ se elimina, lo cual se logra utilizando una relación entre las derivadas de $v$ y $u$. Se obtiene una nueva ecuación en la que la derivada de $v$ respecto a $r$ está relacionada con el valor de $v$ mismo y su derivada. La forma resultante es una ecuación diferencial de primer orden que involucra $v$.

o, después de cierta manipulación,

$$
\frac{d}{dr} \left( r \frac{dv}{dr} - 2v \right) = \frac{1}{2} \frac{d(v^2)}{dr}.
$$

Esta ecuación se obtiene al reorganizar la ecuación anterior, transformándola en una forma más conveniente para la integración. El término $r \frac{dv}{dr}$ se manipula de manera que aparezca una expresión en términos de la derivada de $v^2$. Este es un paso clave en la resolución de la ecuación, ya que transforma la ecuación diferencial en una forma que puede integrarse fácilmente.

__Explicación física:__

Esta manipulación es importante en el contexto de la física porque convierte una ecuación que originalmente involucra derivadas de primer orden en una ecuación relacionada con la cantidad de energía o cantidad conservada (en este caso $v^2$).


Al integrar y reorganizar, obtenemos

$$
r \frac{dv}{dr} = \frac{v (pv + 4)}{2}.
$$

(9.24a)

Al integrar la ecuación obtenida en el paso anterior, se llega a esta expresión en la que $r \frac{dv}{dr}$ está directamente relacionado con $v$ y sus productos. Este es un paso crucial para obtener una solución exacta o aproximada a la ecuación original, permitiendo resolver el comportamiento de $v$ en función de la variable radial $r$. El término $(pv + 4)$ refleja cómo las interacciones entre la temperatura y el parámetro $p$ afectan la evolución de la variable $v$.

__Explicación física:__

Desde el punto de vista físico, esta ecuación describe cómo cambia la cantidad de energía (o alguna propiedad relacionada, como la temperatura) con respecto al radio en un medio. La presencia del término $pv$ indica que el comportamiento de la variable $v$ no es lineal, sino que depende de su propio valor.


El término de integración es cero ya que $v = 0$ cuando $r = 0$. Esta ecuación se puede integrar nuevamente escribiéndola en la forma:
$$ \left( \frac{1}{v} - \frac{p}{pv + 4} \right) dv = 2 \frac{dr}{r}, $$

lo que da:
$$ \log |v| - \log (pv + 4) = 2 \log r + \log A, \quad (9.24b) $$

donde $\log A$ es una constante de integración. Nótese que es claro a partir de la Fig. 9.7 y (9.21) que $v$ es negativo. Por lo tanto, debemos escribir:
$$ \int \frac{dv}{v} = \log |v|. $$

(Este punto a veces no se enfatiza en trabajos elementales). A partir de la forma de $v$ deducida de la Fig. 9.7 y (9.21), es evidente que $dv/dr$ es negativo. De (9.24a), vemos que $pv + 4$ es positivo. La ecuación (9.24b) da:
$$ \frac{|v|}{pv + 4} = Ar^2. $$

Como $v$ es negativo, esto significa que:
$$ \frac{v}{pv + 4} = -Ar^2, $$

y resolviendo para $v$, encontramos que:
$$ v = r \frac{du}{dr} = -\frac{4Ar^2}{1 + pAr^2}, $$

donde $A$ es una constante arbitraria.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parámetros
A = 1.0  # Constante arbitraria
p = 0.5  # Parámetro p
r = np.linspace(0.01, 2, 100)  # Radio (evitar r=0 para evitar singularidad)

# Calcular v(r)
v = -4 * A * r**2 / (1 + p * A * r**2)

# Graficar v contra r
plt.figure(figsize=(8, 6))
plt.plot(r, v, label=r'$v = -\frac{4Ar^2}{1 + pAr^2}$')
plt.title(r'Comportamiento de $v$ en función de $r$')
plt.xlabel('Radio $r$')
plt.ylabel('Valor de $v$')
plt.grid(True)
plt.legend()
plt.show()
```

Integrando nuevamente se obtiene:
$$ u = -\frac{2}{p} \log (1 + pAr^2) + B, $$

donde $B$ es otra constante arbitraria. La condición $u = 0$ en $r = a$ da $B$ en términos de $A$, y encontramos:
$$ u = \frac{2}{p} \log \left( \frac{1 + pAa^2}{1 + pAr^2} \right). \quad (9.25) $$

Para determinar $A$, sustituimos esta solución en la ecuación original (9.20). Encontramos que (9.20) se satisface idénticamente si:
$$ 8A = q(1 + pAa^2)^2. $$



1. La ecuación se integra usando la propiedad de logaritmos. Se resuelve para $v$ usando el término de integración, con la constante arbitraria $A$.
2. Se reescribe la ecuación en términos de $v$ y se obtiene una solución para $v$ en función de $r$.
3. Posteriormente, se integra nuevamente para obtener $u$ en función de $r$.

__Explicación física:__

1. La integración refleja cómo cambia la temperatura ($u$) con la distancia radial $r$ en un sistema térmico.
2. La solución muestra cómo la temperatura varía a lo largo del medio, con la constante $A$ relacionada con las condiciones iniciales.
3. La condición $u = 0$ en $r = a$ asegura que la temperatura se anule en esa posición, lo cual es una condición de frontera común en problemas térmicos.


Si definimos:
$$ pAa^2 = x, \quad \frac{8}{qpa^2} = \alpha, $$

esta ecuación cuadrática en $A$ se convierte en:
$$ x^2 + (2 - \alpha)x + 1 = 0. $$

De la forma de la solución (9.25) es claro que $A$ debe ser real; es decir, $x$ debe ser real, por lo que:
$$ (2 - \alpha)^2 - 4 \geq 0; $$

es decir:
$$ \alpha (\alpha - 4) \geq 0. $$

Como $\alpha > 0$, esto significa que $\alpha \geq 4$, o:
$$ qpa^2 \leq 2. \quad (9.26) $$



```python
# Parámetros
q = 2  # Factor q
p = 1  # Parámetro p
a = 1  # Radio

# Calcular alpha
alpha = q * p * a**2 / 2  # Condición para que la solución sea real

# Graficar alpha con respecto a q
q_vals = np.linspace(0.1, 10, 100)
alpha_vals = q_vals * p * a**2 / 2

plt.figure(figsize=(8, 6))
plt.plot(q_vals, alpha_vals, label=r'$\alpha = \frac{qpa^2}{2}$')
plt.axhline(4, color='r', linestyle='--', label=r'$\alpha = 4$ (Condición crítica)')
plt.title(r'Condición crítica: $\alpha \geq 4$')
plt.xlabel(r'$q$')
plt.ylabel(r'$\alpha$')
plt.grid(True)
plt.legend()
plt.show()
```

Esta es la condición para que la temperatura tienda a un valor finito cuando $t$ tiende a infinito. De lo contrario, la solución tiende al infinito con el tiempo; es decir, ocurre una explosión, como se discutirá más adelante y se muestra diagramáticamente en la Fig. 9.9. Si $qpa^2 = 2$, entonces $x = 1$; es decir, $pAa^2 = 1$, y (9.25) da:
$$ u = \frac{2}{p} \log \left( \frac{2}{1 + (r/a)^2} \right). $$



1. Se define una nueva variable $x = pAa^2$ y un parámetro $\alpha$ para simplificar la ecuación cuadrática que involucra $A$.
2. Se obtiene una condición sobre $\alpha$ para que la solución sea real, es decir, para evitar que $A$ sea complejo.
3. Al imponer que $\alpha \geq 4$, se deduce la condición sobre el parámetro $qpa^2$ para que la temperatura se mantenga finita en el largo plazo.

__Explicación física:__

1. La condición $\alpha \geq 4$ implica que hay un límite en los parámetros físicos ($q$ y $a^2$) para que el sistema térmico no experimente una "explosión" térmica, es decir, para que la temperatura no se dispare a infinito.
2. Si la condición $qpa^2 \leq 2$ se cumple, el sistema alcanza un valor de temperatura finito a largo plazo, y la distribución de temperatura se describe mediante la ecuación dada.
3. El resultado físico muestra cómo la temperatura se distribuye espacialmente cuando el sistema está en equilibrio térmico.


La temperatura en $r = 0$ está dada por:
$$ u = \frac{2 \log 2}{p} \approx \frac{1.39}{p}. \quad (9.27) $$


La ecuación (9.27) muestra el valor de la temperatura $u$ en el centro ($r = 0$) en términos del parámetro $p$. El valor numérico $\log 2 \approx 0.693$ es utilizado para obtener la aproximación de $u \approx \frac{1.39}{p}$.

__Explicación física:__

Esta expresión describe cómo la temperatura en el centro del sistema depende del parámetro $p$, que está relacionado con las propiedades térmicas del material o el sistema en estudio. A medida que $p$ aumenta, la temperatura en el centro disminuye, lo que indica que el sistema se está estabilizando térmicamente.

Ahora abordamos el problema mediante un método aproximado similar al desarrollado en la Sección 9.1. Supongamos que la temperatura en el cuerpo está representada aproximadamente por:
$$ T(r, t) = T_0 + \left[ 1 - \left( \frac{r}{a} \right)^2 \right] Y(t). $$

Esta es una forma aproximada para describir cómo varía la temperatura $T(r, t)$ dentro de un cilindro. Se asume que la temperatura depende de la posición radial $r$ y del tiempo $t$. El término $T_0$ es la temperatura inicial, y $Y(t)$ es una función del tiempo que modula la variación de la temperatura.

__Explicación física:__

Esta aproximación modela la temperatura en un cilindro con una distribución radial de temperatura. A medida que la temperatura varía, $Y(t)$ representa cómo cambia esa distribución con el tiempo.


La ecuación básica está dada por la relación:
$$
\text{[Tasa de aumento de calor total en el cilindro]} = \text{[Tasa total de producción de calor por la reacción]} - \text{[Calor perdido a través de la frontera exterior]}.
$$

La tasa de aumento del calor total en el cilindro, por unidad de longitud, es:
$$ \rho c \frac{dY(t)}{dt} \int_0^a \left[ 1 - \left( \frac{r}{a} \right)^2 \right] 2\pi r \, dr = \frac{1}{2} \pi a^2 \rho c \frac{dY(t)}{dt}. \quad (9.28) $$


La tasa de cambio del calor total en el cilindro se calcula mediante la derivada temporal de $Y(t)$ y la integración de la distribución de temperatura radial en el cilindro.

__Explicación física:__

Este término representa el cambio de energía térmica acumulada en el cilindro debido al cambio de la función $Y(t)$ en el tiempo.


La tasa total de producción de calor por la reacción, por unidad de longitud del cilindro, es:
$$ I = \int_0^a 2\pi Q e^{-P/T} r \, dr. \quad (9.29) $$


Este es el cálculo de la producción de calor interna por la reacción química en el cilindro, integrando sobre el volumen del cilindro con la dependencia radial de $r$.

__Explicación física:__

Este término refleja cómo la reacción química genera calor en función de la distancia radial $r$. La expresión $e^{-P/T}$ describe cómo la producción de calor depende de la temperatura.



Adoptamos la notación y aproximaciones siguiendo (9.18):
$$ Q e^{-P/T} \approx kqe^{P(T - T_0)}. $$

Luego, (9.29) da:
$$ I = 2\pi kqe^Y \int_0^a r e^{-Yr^2/a^2} \, dr = \pi kq a^2 \left( \frac{e^Y - 1}{pY} \right). \quad (9.30) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parámetros
k = 1.0    # Constante de conductividad térmica
q = 1.0    # Constante relacionada con la producción de calor
a = 1.0    # Radio del cilindro
p = 0.5    # Parámetro de reacción

# Rango de valores para Y
Y = np.linspace(0.01, 5, 100)

# Calcular I(Y)
I = np.pi * k * q * a**2 * (np.exp(Y) - 1) / (p * Y)

# Graficar I contra Y
plt.figure(figsize=(8, 6))
plt.plot(Y, I, label=r'$I = \pi k q a^2 \frac{e^Y - 1}{p Y}$')
plt.title('Tasa de Producción de Calor $I$ en función de $Y$')
plt.xlabel('Temperatura $Y$')
plt.ylabel('Producción de Calor $I$')
plt.grid(True)
plt.legend()
plt.show()
```



Se hace una aproximación de la forma de $Q e^{-P/T}$, resultando en una función que depende de $Y(t)$. La integral refleja la contribución de la producción de calor a lo largo del radio del cilindro.

__Explicación física:__

El término $e^Y$ indica cómo la producción de calor aumenta con el tiempo debido a la reacción química. La expresión $(e^Y - 1)/pY$ muestra cómo esta producción se distribuye en función de la temperatura $Y$.



La tasa de pérdida de calor a través de la frontera $r = a$ es:
$$ -2\pi ak \left( \frac{dT}{dr} \right)_{r=a} = 4\pi k Y. \quad (9.31) $$


Este término describe la pérdida de calor en la frontera exterior del cilindro, donde la derivada radial de la temperatura $T$ se evalúa en $r = a$.

```python
# Calcular la tasa de pérdida de calor en la frontera
heat_loss = 4 * np.pi * k * Y

# Graficar la tasa de pérdida de calor contra Y
plt.figure(figsize=(8, 6))
plt.plot(Y, heat_loss, color='orange', label=r'Pérdida de Calor $= 4 \pi k Y$')
plt.title('Tasa de Pérdida de Calor en función de $Y$')
plt.xlabel('Temperatura $Y$')
plt.ylabel('Pérdida de Calor')
plt.grid(True)
plt.legend()
plt.show()
```

__Explicación física:__

La ecuación muestra cómo el calor se pierde a través de la superficie exterior del cilindro, proporcional a la variación de la temperatura en esa frontera.


La ecuación de equilibrio térmico es, por lo tanto, a partir de (9.28), (9.30) y (9.31):
$$ \frac{1}{2} \pi a^2 \rho c \frac{dY}{dt} = kqa^2 \left( \frac{e^Y - 1}{pY} \right) - 4kY. $$

Si definimos:
$$ pY = Z, \quad \frac{4}{qpa^2} = \beta, \quad \frac{\rho c}{2kqp} = \gamma, $$

esta ecuación se convierte en:
$$ \gamma \frac{dZ}{dt} = \frac{e^Z - 1}{Z} - \beta Z = F(Z), \quad (9.32) $$

donde esto define $F(Z)$.

Se definen nuevas variables para simplificar la ecuación y obtener una forma más manejable, que es una ecuación diferencial en términos de $Z$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de Z y beta
Z = np.linspace(0, 3.5, 100)  # Rango de Z
beta_values = [1.0, 1.5, 2.0]  # Valores de beta como en el gráfico

# Definir F(Z) ajustada para iniciar en F(0) = 1
plt.figure(figsize=(8, 6))
for beta in beta_values:
    F_Z = np.exp(Z) - 1 - beta * Z + 1  # Ajuste para que F(0) = 1
    plt.plot(Z, F_Z, label=r'$\beta = {:.1f}$'.format(beta))

# Configuración del gráfico
plt.axhline(0, color='black', linestyle='--')  # Línea de F(Z) = 0
plt.axhline(1, color='gray', linestyle='--', label='$F(Z) = 1$')  # Línea de F(Z) = 1
plt.title('Función $F(Z)$ como función de $Z$ para varios valores de $\\beta$')
plt.xlabel('$Z$')
plt.ylabel('$F(Z)$')
plt.ylim(-1, 6)  # Ajuste de límites para observar el mínimo de las curvas
plt.grid(True)
plt.legend()
plt.show()
```



__Explicación física:__

La ecuación (9.32) modela la dinámica del sistema térmico de forma más directa, con la función $F(Z)$ describiendo cómo las tasas de producción y pérdida de calor afectan la evolución temporal de $Z$.



La derecha de esta ecuación diferencial, $F(Z)$, se grafica como función de $Z$ para varios $\beta$ en la Fig. 9.8. Se observa que para $\beta$ menor que un valor crítico $\beta_0$, la función $F(Z)$ siempre es positiva. Esto significa que, a partir de la ecuación diferencial, $Z$ tenderá al infinito conforme $t$ tiende al infinito; es decir, ocurre una explosión.

La función $F(Z)$ determina el comportamiento de $Z$ en función de su valor actual. Para ciertos valores de $\beta$, $F(Z)$ siempre es positiva, lo que lleva a una solución explosiva.

__Explicación física:__

Cuando la función $F(Z)$ es positiva, $Z$ crece indefinidamente, lo que implica un aumento descontrolado de la temperatura, conocido como "explosión térmica". Esto ocurre cuando el valor de $\beta$ es suficientemente pequeño, lo que describe una inestabilidad térmica en el sistema.

El valor crítico $\beta_0$ ocurre para un valor de $Z$ tal que $F(Z) = F'(Z) = 0$:

$$
\frac{e^Z - 1}{Z} - \beta_0 Z = 0, \quad -\frac{e^Z}{Z^2} + \frac{e^Z}{Z} - \beta_0 = 0.
$$

Estas ecuaciones definen el valor crítico de $Z$ donde la función $F(Z)$ y su derivada son cero. Es la condición para que el sistema pase de un comportamiento estable a inestable.

__Explicación física:__

Este es el punto crítico donde la temperatura comienza a diverger (explosión térmica).


Al eliminar $\beta_0$, encontramos que $Z$ satisface

$$
e^Z = \frac{2}{2 - Z}.
$$

Al despejar $\beta_0$, obtenemos una ecuación no lineal que se puede resolver numéricamente para encontrar $Z$.

__Explicación física:__

Esta ecuación determina el valor crítico de $Z$ donde el sistema alcanza el punto de explosión térmica.



La solución numérica de esta ecuación da, aproximadamente, $Z = 1.59$. El valor correspondiente de $\beta_0$ está dado por

$$
\beta_0 = \frac{1}{Z (2 - Z)} \approx 1.54.
$$

Se obtiene $Z \approx 1.59$ y $\beta_0 \approx 1.54$ al resolver la ecuación numéricamente.

__Explicación física:__

Estos valores indican el umbral crítico a partir del cual el sistema puede experimentar una explosión térmica.

Debería quedar claro ahora que las soluciones de (9.32) se comportan cualitativamente como se muestra en la Figura 9.9. Esto ilustra una de las ventajas de la teoría simple: la visión directa que proporciona sobre la forma de las soluciones.

Los temas matemáticos clave para resolver estos problemas son:

1. __Ecuaciones Diferenciales Parciales (EDP):__

Las ecuaciones de difusión con fuentes no lineales son un tema central. Por ejemplo, la ecuación (9.19) describe cómo la temperatura cambia en función del tiempo y la posición radial, con términos no lineales dependientes de la temperatura, como $e^{-P/T}$. La ecuación (9.20), que describe el estado estacionario del sistema, se convierte en una ecuación diferencial ordinaria (EDO) no lineal de segundo orden en $r$.

2. __Expansiones en Serie y Aproximaciones Lineales:__

Se utiliza la expansión binomial de primer orden para aproximar términos como
$$
\left( 1 + \frac{T - T_0}{T_0} \right)^{-1},
$$
que permite simplificar las ecuaciones y hacerlas tratables sin perder precisión significativa en rangos pequeños de variación de temperatura $T$. Estas aproximaciones son fundamentales cuando el sistema está cerca de un valor de referencia (en este caso $T_0$).

3. __Métodos de Resolución de Ecuaciones No Lineales:__

El texto presenta una serie de transformaciones y cambios de variable que permiten simplificar la ecuación original y convertirla en una forma más manejable, como la ecuación (9.22) donde se define
$$
v = r \frac{du}{dr},
$$
y la ecuación (9.24a), que es más fácil de resolver. También se utiliza la separación de variables en algunos momentos, como en la integración de la ecuación diferencial (9.24b), lo que permite una solución analítica para $v$. Además, se utilizan integrales de la forma
$$
\int_0^a r e^{-Y r^2/a^2} \, dr
$$
para obtener las soluciones de producción de calor y de pérdida de calor en el sistema.

4. __Condiciones de Frontera y Estado Estacionario:__

La condición
$$
u = 0 \quad \text{en} \quad r = a
$$
es clave para encontrar soluciones específicas que cumplen con las condiciones físicas del problema (es decir, que la diferencia de temperatura se anule en la frontera del sistema). El estado estacionario (cuando
$$
\frac{\partial u}{\partial t} = 0
$$
es crucial, ya que permite reducir las ecuaciones diferenciales a formas que son más fáciles de resolver y analizar.

5. __Análisis de Estabilidad:__

El valor crítico de $\beta_0$, que separa el comportamiento de las soluciones entre estabilidad e inestabilidad, se obtiene mediante el análisis de la función $F(Z)$. Este tipo de análisis es fundamental para comprender el comportamiento de las soluciones a largo plazo, especialmente en sistemas térmicos que pueden experimentar explosiones térmicas si las condiciones no se controlan adecuadamente.

6. __Aproximaciones Exponenciales:__

Las soluciones de tipo
$$
e^{-P/T}
$$
son comunes en problemas donde las reacciones químicas o los procesos de difusión están involucrados. Las aproximaciones como
$$
Q e^{-P/T} \approx k q e^{P(T - T_0)}
$$
son esenciales para transformar la ecuación no lineal en algo más simple, permitiendo el análisis de la solución en términos de las variaciones de temperatura.
